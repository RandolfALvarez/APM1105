```python
import requests

# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')
```

```python
#NUMBER 1 A AND B
fname = "words.txt"
 
num_words = 0
 
with open(fname, 'r') as f:
    for line in f:
        words = line.split()
        num_words += len(words)
print("A) Number of words:",num_words)


with open(fname, 'r') as filedata:
    wordsList = filedata.read().split()
longestWordLength = len(max(wordsList, key=len))
result = [textword for textword in wordsList if len(textword) == longestWordLength]
print("B) Longest word/s:",result)
```

```python
#NUMBER 1 E
def is_palindrome(word):
  
    return word == word[::-1]


def filepalindromes(file_path):
  
    with open(file_path, 'r') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                if is_palindrome(word):
                    print(word)
                    
fname='words.txt'

filepalindromes(fname)
```

```python
#NUMBER 1 G
import random

def game():
    
    fname='words.txt'
    with open(fname, "r") as file:
        allText = file.read()
        words = list(map(str, allText.split()))

    word = random.choice(words)
    guesses = []
    lives = 10
    
    while lives > 0:
        display_word = ""
        for letter in word:
            if letter in guesses:
                display_word += letter
            else:
                display_word += "_"
        print(display_word)
        
        guess = input("Guess a letter: ")
 
        if guess in guesses:
            print("You already guessed that letter!")
        else:
            guesses.append(guess)
            if guess in word:
                print("Correct!")
            else:
                print("Incorrect!")
                lives -= 1
        
        if set(word) <= set(guesses):
            print("Congratulations, you guessed the word!")
            return
        

        print(f"You have {lives} lives remaining.")
    
    print("Game over. The word was", word)
```

```python
#NUMBER 2
import time

def fibonacci_iterative(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        fib1 = 0
        fib2 = 1
        for i in range(2, n+1):
            fib = fib1 + fib2
            fib1 = fib2
            fib2 = fib
        return fib

def fibonacci_recursive(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

n = 30
start_time = time.time()
fib_iter = fibonacci_iterative(n)
end_time_iter = time.time()
print(f"Fibonacci sequence (iterative) for n={n}: {fib_iter}")
print(f"Rendering time (iterative): {end_time_iter - start_time}")

start_time = time.time()
fib_recur = fibonacci_recursive(n)
end_time_recur = time.time()
print(f"Fibonacci sequence (recursive) for n={n}: {fib_recur}")
print(f"Rendering time (recursive): {end_time_recur - start_time}")
```

```python
#NUMBER3
def formula(x):
    return pow(20 - 5*x, 1/3)

x = 2

print(f"x_0 = {x:.3f}")

tolerance = 0.001
while True:
    x_next = formula(x)
    print(f"x_{int(x/x)+1} = {x_next:.3f}")
    if abs(x_next - x) < tolerance:
        break
    x = x_next

print(f"The solution to x^3 + 5x = 20 is: {x_next:.3f}")
```
